#### –≠—Ç–æ –ø–µ—Ä–≤–∏—á–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∫–æ—Ç–æ—Ä—ã–µ –Ω–∞–º –≤—ã–¥–∞–ª–∏ –¥–ª—è —Ä–∞–∑–º–µ—Ç–∫–∏

```python
import pandas as pd

df = pd.read_csv("123.csv")
df
```

#### –†–∞–∑–º–µ—á–µ–Ω–Ω—ã–µ –º–Ω–æ—é –¥–∞–Ω–Ω—ã–µ

##### –≥–¥–µ-—Ç–æ –ø–æ—Ç–µ—Ä—è–ª–æ—Å—å 5 –∫–æ–º–µ–Ω—Ç–∞—Ä–∏–µ–≤

```python
dt = pd.read_csv("321.csv")
dt
```

#### –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—É—é –∫–æ–ª–æ–Ω–∫—É –∏ –∑–ø–æ–ª–Ω—è–µ–º –µ–µ 0

```python
dt['–Ω–∏ –∫ –æ–¥–Ω–æ–º—É –∏–∑ –ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö'] = 0
dt
```

#### –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–µ—Ä–µ–Ω–æ—Å–∞ —Ä–∞–∑–º–µ—Ç–∫–∏ –≤ —Å–≤–æ–∏ –∫–æ–ª–æ–Ω–∫–∏ –∏–∑ –∫–æ–ª–æ–Ω–∫–∏ 'sentiment'

```python
import ast
stolbiki = [
    '–í–æ–ø—Ä–æ—Å —Ä–µ—à–µ–Ω',
    '–ù—Ä–∞–≤–∏—Ç—Å—è –∫–∞—á–µ—Å—Ç–≤–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞—è–≤–∫–∏',
    '–ù—Ä–∞–≤–∏—Ç—Å—è –∫–∞—á–µ—Å—Ç–≤–æ —Ä–∞–±–æ—Ç—ã —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤',
    '–ù—Ä–∞–≤–∏—Ç—Å—è —Å–∫–æ—Ä–æ—Å—Ç—å –æ—Ç—Ä–∞–±–æ—Ç–∫–∏ –∑–∞—è–≤–æ–∫',
    '–ü–æ–Ω—Ä–∞–≤–∏–ª–æ—Å—å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –∑–∞—è–≤–∫–∏',
    '–Ω–∏ –∫ –æ–¥–Ω–æ–º—É –∏–∑ –ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö'
]

for idx, row in dt.iterrows():
    try:
        sentiment = row['sentiment']
        if isinstance(sentiment, str) and '"choices"' in sentiment:
            data = ast.literal_eval(sentiment)
            for label in data.get("choices", []):
                if label in stolbiki:
                    dt.at[idx, label] = 1
        elif sentiment in stolbiki:
            dt.at[idx, sentiment] = 1
    except Exception as e:
        print(f"–û—à–∏–±–∫–∞ –≤ —Å—Ç—Ä–æ–∫–µ {idx}: {e}")

dt
```

#### –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è 0 —á—Ç–æ–±—ã –Ω–µ–±—ã–ª–æ –ø—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ –Ω–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö

```python
colum = ['–í–æ–ø—Ä–æ—Å —Ä–µ—à–µ–Ω',
    '–ù—Ä–∞–≤–∏—Ç—Å—è –∫–∞—á–µ—Å—Ç–≤–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –∑–∞—è–≤–∫–∏',
    '–ù—Ä–∞–≤–∏—Ç—Å—è –∫–∞—á–µ—Å—Ç–≤–æ —Ä–∞–±–æ—Ç—ã —Å–æ—Ç—Ä—É–¥–Ω–∏–∫–æ–≤',
    '–ù—Ä–∞–≤–∏—Ç—Å—è —Å–∫–æ—Ä–æ—Å—Ç—å –æ—Ç—Ä–∞–±–æ—Ç–∫–∏ –∑–∞—è–≤–æ–∫',
    '–ü–æ–Ω—Ä–∞–≤–∏–ª–æ—Å—å –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–µ –∑–∞—è–≤–∫–∏',
    '–Ω–∏ –∫ –æ–¥–Ω–æ–º—É –∏–∑ –ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö']
dt[colum] = dt[colum].fillna(0)
dt
```

#### –°–º–æ—Ç—Ä–∏–º –∫–∞–∫ —Ä–∞–∑–º–µ—á–µ–Ω—ã –Ω–∞—à–∏ –¥–∞–Ω–Ω—ã–µ —Å–∫–æ–ª—å –∫–æ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –∫–∞–∂–¥–æ–º –∑–Ω–∞—á–µ–Ω–∏–∏

```python
summa = dt[colum].sum()
summa
```

#### –°–¥–µ–ª–∞—Ç—å –∫–∞—Ä–µ–∫—Ç–Ω–æ–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–æ–ª–æ–Ω–∫–∏–∏

–£–¥–∞–ª–µ–Ω–∏–µ –ª–∏—à–∫–∏—Ö –∫–æ–ª–æ–Ω–æ–∫ –Ω–∞ –ø–µ—Ä–≤–æ–º —ç—Ç–∞–ø–µ üêõ

```python
dt = dt.rename(columns={'–Ω–∏ –∫ –æ–¥–Ω–æ–º—É –∏–∑ –ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω—ã—Ö': '–ü—Ä–æ–±–ª–µ–º—ã'})
dt = dt.drop(['created_at', 'updated_at', 'annotation_id', 'annotator', 'sentiment', 'lead_time','id'], axis=1)
dt
```
